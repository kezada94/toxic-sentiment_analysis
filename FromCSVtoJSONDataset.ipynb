```python
import pandas as pd
import json
import numpy as np
import torch
```

```python
train_csv=pd.read_csv("data/train.csv")
test_csv=pd.read_csv("data/test.csv")
test_lcsv = pd.read_csv("data/test_labels.csv")
print(train_csv.head(7))
print(test_csv.head(7))
test_lcsv.head(7)
```

```python
train_json={}
train_json['comment_text']=train_csv['comment_text'].values.tolist()
train_json['toxic']=train_csv.iloc[:, 2].values.tolist()
train_json['severe_toxic']=train_csv.iloc[:, 3].values.tolist()
train_json['obscene']=train_csv.iloc[:, 4].values.tolist()
train_json['threat']=train_csv.iloc[:, 5].values.tolist()
train_json['insult']=train_csv.iloc[:, 6].values.tolist()
train_json['identity_hate']=train_csv.iloc[:, 7].values.tolist()

test_json={}
test_json['id']=test_csv['id'].values.tolist()
test_json['comment_text']=test_csv['comment_text'].values.tolist()
test_json['toxic']=test_lcsv.iloc[:, 1].values.tolist()
test_json['severe_toxic']=test_lcsv.iloc[:, 2].values.tolist()
test_json['obscene']=test_lcsv.iloc[:, 3].values.tolist()
test_json['threat']=test_lcsv.iloc[:, 4].values.tolist()
test_json['insult']=test_lcsv.iloc[:, 5].values.tolist()
test_json['identity_hate']=test_lcsv.iloc[:, 6].values.tolist()

```

```python
#lista=[]
trainvalid = np.random.choice(len(train_csv), int(len(train_csv)*0.2))
for i, data in enumerate(train_json['comment_text']):
    jot={}
    jot["comment_text"]=data
    jot["toxic"]=train_json['toxic'][i]
    jot["severe_toxic"]=train_json['severe_toxic'][i]
    jot["obscene"]=train_json['obscene'][i]
    jot["threat"]=train_json['threat'][i]
    jot["insult"]=train_json['insult'][i]
    jot["identity_hate"]=train_json['identity_hate'][i]
    if i in trainvalid:
        with open('data/valid.json', 'a') as f:
            json.dump(jot, f)
            f.write("\n")
    else:
        with open('data/train.json', 'a') as f:
            json.dump(jot, f)
            f.write("\n")

for i,data in enumerate(test_json['comment_text']):
    if test_lcsv['toxic'][i] == -1:
        continue
    jot={}
    jot["comment_text"]=data
    jot["toxic"]=test_json['toxic'][i]
    jot["severe_toxic"]=test_json['severe_toxic'][i]
    jot["obscene"]=test_json['obscene'][i]
    jot["threat"]=test_json['threat'][i]
    jot["insult"]=test_json['insult'][i]
    jot["identity_hate"]=test_json['identity_hate'][i]    
    with open('data/test.json', 'a') as f:
        json.dump(jot, f)
        f.write("\n")
    

    
```
